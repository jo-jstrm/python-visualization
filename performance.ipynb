```python
import pandas as pd
import seaborn as sns
```

```python
data_small_workloads = data[data['Workload Size'] < 134217728] 
data_large_workload = data[data['Workload Size'] == 134217728]
# print(data_small_workloads)
# print(data_large_workload)
```

```python
sns.set(rc={"figure.figsize":(10, 10)}) # sets the theme to the default seaborn theme
sns.set_theme()
sns.barplot(x=data_large_workload['Workgroup Size'],
             y=data_large_workload['Timing'])
```
